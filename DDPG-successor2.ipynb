```python
import numpy as np
import stable_baselines3
from stable_baselines3 import DDPG
from stable_baselines3.common.noise import NormalActionNoise, OrnsteinUhlenbeckActionNoise
```

```python
n_actions = env.action_space.shape[-1]
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.5 * np.ones(n_actions))

tensorboard_log = os.path.join(logdir, "tensorboard")
model = DDPG("MlpPolicy", env, action_noise=action_noise, verbose=1, batch_size=256, gamma=0.1, learning_rate=0.0001, tensorboard_log=tensorboard_log)
TIMESTEPS = 30000000
for i in range(1,30):
    model.learn(total_timesteps=TIMESTEPS, log_interval=10, progress_bar=True, reset_num_timesteps=False, tb_log_name="DDPG-successor2")
    model.save(f"{models_dir}/{TIMESTEPS*i}")
env = model.get_env()
```
