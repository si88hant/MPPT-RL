```python
import gym
import gym_mppt
from gym import envs

import os
models_dir = "models/TD3"
logdir = "logs"

if not os.path.exists(models_dir):
    os.makedirs(models_dir)
    
if not os.path.exists(logdir):
    os.makedirs(logdir)    

env = gym.make("mppt_shaded-v1")

num_states = env.observation_space.shape[0]
print("Size of State Space ->  {}".format(num_states))
num_actions = env.action_space.shape[0]
print("Size of Action Space ->  {}".format(num_actions))

upper_bound = env.action_space.high[0]
lower_bound = env.action_space.low[0]

print("Max Value of Action ->  {}".format(upper_bound))
print("Min Value of Action ->  {}".format(lower_bound))
```

```python
import numpy as np
from stable_baselines3 import TD3
from stable_baselines3.common.noise import NormalActionNoise, OrnsteinUhlenbeckActionNoise
```

```python
n_actions = env.action_space.shape[-1]
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))

tensorboard_log = os.path.join(logdir, "tensorboard")
model = TD3("MlpPolicy", env, action_noise=action_noise, verbose=1, learning_rate=0.000001, policy_delay=5, gamma=0.20, tensorboard_log=tensorboard_log)
TIMESTEPS = 20000000
for i in range(1,30):
    model.learn(total_timesteps=TIMESTEPS, log_interval=10, progress_bar=True, reset_num_timesteps=False, tb_log_name="TD3-successor1")
    model.save(f"{models_dir}/{TIMESTEPS*i}")
env = model.get_env()
```
